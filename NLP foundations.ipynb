# Checking English tuples

```python
def is_english(text):
    for lang in ['en', 'english']:
        if lang in text.lower():
            return True
    return False

# Apply the function to the tweets column
df['is_english'] = df['text'].apply(is_english)

# Drop rows where the tweets are not in English
df = df[df['is_english'] == True]

# Print the dataframe
print(df['text'])
```

# Tokenization

**WhitespaceTokenizer**

```python
from nltk.tokenize import WhitespaceTokenizer
Wspace = WhitespaceTokenizer()
df_text_string = df.text.to_string()
print(Wspace.tokenize(df_text_string))
```

**WordPunctTokenizer**

```python
from nltk.tokenize import WordPunctTokenizer
wpt=WordPunctTokenizer()
print("Punctuation Based tokenization = ",wpt.tokenize(df_text_string))
```

**TreebankWordTokenizer**

```python
from nltk.tokenize import TreebankWordTokenizer
tb = TreebankWordTokenizer()
print("Treebank Tokenizer =",tb.tokenize(df_text_string))
```

**TweetTokenizer**

```python
from nltk.tokenize import TweetTokenizer
tt= TweetTokenizer()
print("Tweet Tokenizer =",tt.tokenize(df_text_string))
```

**MWETokenizer**

```python
from nltk.tokenize import MWETokenizer
mwet = MWETokenizer([('Chat', 'GPT')])
print("MWE Tokenizer =",mwet.tokenize(tt.tokenize(df_text_string)))
```

# STEMMING

**PorterStemmer**

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
ps = PorterStemmer()
words= word_tokenize(df_text_string)
for word in words:
  print(word, " : ", ps.stem(word))
```

**SnowballStemmer**

```python
from nltk.stem import SnowballStemmer
from nltk.tokenize import word_tokenize
ss = SnowballStemmer(language="english")
words= word_tokenize(df_text_string)
for word in words:
  print(word, " : ", ss.stem(word))
```

**LancasterStemmer**

```python
from nltk.stem import LancasterStemmer
from nltk.tokenize import word_tokenize
ls = LancasterStemmer()
words= word_tokenize(df_text_string)
for word in words:
  print(word, " : ", ls.stem(word))
```

**Comparison between all 3 implemented stemmers**

```python
# example 1
print("PorterStemmer: " + ps.stem("children"))
print("SnowballStemmer: " + ss.stem("children"))
print("LancasterStemmer: " + ls.stem("children"))

# example 2
print("PorterStemmer: " + ps.stem("older"))
print("SnowballStemmer: " + ss.stem("older"))
print("LancasterStemmer: " + ls.stem("older"))
```

# Lemmatization

**NLTK Lemmatization**

```python
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
words = nltk.word_tokenize(df_text_string)
for word in words:
  print(word, " : ", wnl.lemmatize(word))
```

**spaCy Lemmatization**

```python
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp(df_text_string)
for word in doc:
  print(word, " : ",word.lemma_)
```

# Part of speech tagging

**Stochastic POS Tagging**

```python
from nltk.tokenize import word_tokenize
words= word_tokenize(df_text_string)
tags = nltk.pos_tag(words)
for word, tag in tags:
    print(word+ " : " + tag)
```

**Transformer-Based POS Tagging**

```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp(df_text_string)
for token in doc:
    print(token.text +" : "+ token.pos_)
```
